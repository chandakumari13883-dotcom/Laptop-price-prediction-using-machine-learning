```python
import numpy as np
import pandas as pd
```

```python
#2.data loading
df=pd.read_csv('/laptop_data.csv',index_col=0)
```

```python
#3.BASIC EDA (Exploratory Data Analysis)
df.head()
```

```python
df.describe()
```

```python
#4.Data Cleaning
#ex- remove unwanted columns,fix data types,handle missing values
df.drop(columns=['Unnamed:0'],inplace=True,errors='ignore')
```

```python
df.columns
```

```python
#convert ram to numric
if df['Ram'].dtype=='object':
    df['Ram']=df['Ram'].str.replace('GB','')
    df['Ram']=df['Ram'].astype('int32')
```

```python
#convert weight to numeric
if df['Weight'].dtype=='object':
    df['Weight']=df['Weight'].str.replace('kg','')
    df['Weight']=df['Weight'].astype('float32')
```

```python
x=df.drop('Price',axis=1)
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
categorical_cols=x.select_dtypes(include='object').columns
preprocessor=ColumnTransformer(transformers=[('cat',OneHotEncoder(drop='first',handle_unknown='ignore'),categorical_cols)],remainder='passthrough')
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression

model= Pipeline(steps=[('preprocessor',preprocessor),('regresson',LinearRegression())])
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
model.fit(x_train,y_train)
```
